```python
ipl_scores = [200, 150, 300, 250, 400, 165, 350, 275, 500, 100, 450, 225, 375, 125, 425]
#statistical calculations on the list
print(f"Scores: {ipl_scores}")
print(f"Total: {sum(ipl_scores)}")
print(f"Average: {sum(ipl_scores) / len(ipl_scores):.2f}")
print(f"Maximum: {max(ipl_scores)}")
print(f"Minimum: {min(ipl_scores)}")
print(f"Sorted Scores: {sorted(ipl_scores)}")
print(f"Reverse Sorted Scores: {sorted(ipl_scores, reverse=True)}")
print(f"Count of 250: {ipl_scores.count(250)}")
print(f"length of the list: {len(ipl_scores)}")
top3 = sorted(ipl_scores, reverse=True)[:3]
print(f"Top 3 scores: {top3}")
```

```python
original = [1,2,3]
wrong_copy = original # this is not a copy, it is just another reference to the same list, No chanes in memory, 
#so if we change the original list, the wrong_copy will also change

right_copy = original.copy() # this is a copy, it is a new list in memory, ---> always use copy() method to create a copy of the list,
#so if we change the original list, the right_copy will not change

right_copy.append(4) # this will add 4 to the right_copy list, but not to the original list
print(f"Original List: {original}")
print(f"Right Copy List: {right_copy}")
```
