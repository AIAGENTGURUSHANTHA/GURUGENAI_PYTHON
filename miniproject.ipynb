```python
#Features:
# 1. Start with an empty watchlist
watchlist = []

#2. 
watched = []

while True:
    print("Watchlist Manager")
    print("--" * 30)
    print("1. Add a movie to the watchlist")
    print("2. Remove a movie from the watchlist")
    print("3. Mark a movie as watched")
    print("4. Show All")
    print("5. Show Stats")
    print("6. Sort Watchlist")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    #Add movie to the watchlist
    if choice == "1":
        movie = input("Enter the name of the movie to add: ").lower()
        if movie not in watchlist and movie not in watched:
            watchlist.append(movie)
            print(f"{movie} Movie has been added to your watchlist successfully.")
        else:
            print(f"{movie} Movie is already in your watchlist or watched list.")
    
    # Remove movie from the watchlist
    elif choice == "2":
        movie = input("Enter the name of the movie to remove: ").lower()
        if movie in watchlist:
            watchlist.remove(movie)
            print(f"{movie} Movie has been removed from your watchlist successfully.")
        else:
            print(f"{movie} Movie is not in your watchlist.")
    # Mark movie as watched
    elif choice == "3":
        movie = input("Enter the name of the movie to mark as watched: ").lower()
        if movie in watchlist:
            watchlist.remove(movie)
            watched.append(movie)
            print(f"{movie} Movie has been updated in watched successfully.")
        else:
            print(f"{movie} Movie is not found in your watchlist.")
    
    # Show all movies in the watchlist and watched list
    elif choice == "4":
        print("Your Watchlist:")
    else:
        break
    

    
```
