Variable: which holds the value in it.
Text -->"" -> string --> Text
whole number -> integer -->int
decimal number -> float -> float
Boolean -> True/False -> Yes/No -> bool


```python
#variables
player_name = "Virat Kohli"  #Text --- string(str)
age = 36 #whole number --- integer(int)
batting_average = 58.16 #decimal number --- float(float)
is_captain = False #True or False --- boolean(bool)
jersey_number = 18 #whole number --- integer(int)

print("================================")
print("     Player Profile Card")
print("================================")   #total number of equal signs = 32
print("Name :",player_name, age) # to print variable value -> we need to use print() function without quotes
print("Age :",age)
print("Batting Average :",batting_average)
print("Is Captain :",is_captain)
print("Jersey Number :",jersey_number)
print("================================")
```
