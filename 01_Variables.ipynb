```python
#variables
player_name = "Virat Kohli"  #Text --- string(str)
age = 36 #whole number --- integer(int)
batting_average = 58.16 #decimal number --- float(float)
is_captain = False #True or False --- boolean(bool)
jersey_number = 18 #whole number --- integer(int)

print("================================")
print("     Player Profile Card")
print("================================")   #total number of equal signs = 32
print("Player Name      : ",player_name)
print("Age              : ",age)
print("Batting Average  : ",batting_average)
print("Is Captain       : ",is_captain)
print("Jersey Number    : ",jersey_number)
print("================================")
```

```python
#string (str) data type -> any text, always enclosed in single or double quotes
player_name = "Rohit Sharma"  #string(str)
team_name = 'India'  #string(str) # single quotes or double quotes can be used to define a string
nickname = "Hitman"  #string(str)
home_ground = "Wankhede Stadium"   #string(str)

#integer (int) data type -> whole numbers, no decimal point
matches_played = 227  #integer(int)
centuries_scored = 43  #integer(int)
sixes_hit = 205  #integer(int)
jersey_number = 45  #integer(int)

#float (float) data type -> decimal numbers, can have a decimal point
batting_average = 48.96  #float(float)
strike_rate = 88.34  #float(float)
economy_rate = 5.12  #float(float)

#Boolean (bool) data type -> True or False values
is_captain = True  #boolean(bool)
is_injured = False  #boolean(bool)
is_overseas_player = False  #boolean(bool)

#none (NoneType) data type -> represents the absence of a value or a null value -> 
#nothing, empty, null, not decided, not assigned, not initialized
retirement_date = None  #none(NoneType)

#type is a keyword in Python that is used to determine the data type of a variable or value.
#type () function returns the type of the object passed to it as an argument.
#type() tells us what type of data a variable is holding, such as string, integer, float, boolean, etc.

print("================================")
print("Player:", player_name,  "|  type:", type(player_name))
print("Matches Played:", matches_played,  "|  type:", type(matches_played))
print("Batting Average:", batting_average,  "|  type:", type(batting_average))
print("Is Captain:", is_captain,  "|  type:", type(is_captain))
print("Retirement Date:", retirement_date,  "|  type:", type(retirement_date))
print("================================")

#variable naming rules in Python:
#1. Variable names can only contain letters (a-z, A-Z), numbers (0-9), and underscores (_). They cannot contain spaces or special characters.
#2. Variable names must start with a letter or an underscore (_). They cannot start with a number.
#3. Variable names are case-sensitive, meaning that uppercase and lowercase letters are treated as different characters. 
# For example, "player_name" and "Player_Name" would be considered two different variables.
#snake_case: player_name, batting_average, is_captain
#pascalCase: PlayerName, BattingAverage, IsCaptain
#camelCase: playerName, battingAverage, isCaptain

#type will be identified by the type() function based on the value assigned to the variable. 
# Python is a dynamically typed language, which means that you don't need to explicitly declare the data type of a variable when you create it. The interpreter infers the type based on the value assigned to the variable.
```
